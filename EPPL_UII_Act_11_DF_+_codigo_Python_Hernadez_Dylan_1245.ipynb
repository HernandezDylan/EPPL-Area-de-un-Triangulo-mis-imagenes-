```python
# Calcular Area de un triangulo rectangulo
# Variables A, b, h
# Formula A = (b * h) / 2

print("Calcular Area de un triangulo rectangulo")

b = float(input("Ingrese la base: "))
h = float(input("Ingrese la altura: "))

A = (b * h) / 2

print("El area del triangulo es: ", A)
print("Hernadez Dylan NL = 24")
```
